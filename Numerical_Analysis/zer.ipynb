In the powerpoint, the Zernike Polynomials are defined as...</br>
$Z_n^m(\rho, \theta) = N_n^mR_n^{|m|}(\rho)cos(m\theta)$      for $m\geq 0$, $0\leq p\leq 1$, $0\leq\theta\leq 2\pi$ </br>
</br>
and </br>
</br>
$Z_n^m(\rho, \theta) = -N_n^mR_n^{|m|}(\rho)sin(m\theta)$      for $m < 0$, $0\leq p\leq 1$, $0\leq\theta\leq 2\pi$
</br>
</br>
for a given $n$: $m$ can take values of $-n, -n+2, -n+4,...,n$</br>
</br>
$N_n^m$ is the normalization factor</br>
</br>
<p style="font-size:18px">$N_n^m = \sqrt{\frac{2(n+1)}{1+\delta_{m0}}}$</p></br>
$\delta_{m0}$ is dependent on $m$.</br>
$\delta_{m0} = 1$ for $m=0$, $\delta_{m0} = 0$ for $m\neq 0$

$R_n^{|m|}(\rho)$ is the radial polynomial</br>
$R_n^{|m|}(\rho) = \sum_{s=0}^{(n-|m|)/2}$ $\frac{(-1)^s(n-s)!}{s![0.5(n+|m|)-s]![0.5(n-|m|)-s]!}\rho^{n-2s}$

```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import matplotlib.pyplot as plt
import numpy as np
import math

class Zernike():
    def __init__(self):
        self.theta_vals = np.linspace(0, 2*np.pi, 100)
        self.rho_vals = np.linspace(0, 1, 100)
    
    def get_N(self, n, m):
        """
        finds and returns the normalization factor.  
        Also makes self.N = normalization factor
        """
        if m == 0:
            delta = 1
            N = np.sqrt((2*(n + 1))/(1 + delta))
        else:
            delta = 0
            N = np.sqrt((2*(n + 1))/(1 + delta))
        self.N = N
        return N
        
    def R_root(self, n, m, rho, s):
        """
        This is the computation for the inside of the summation.
        The actual summation is done in get_R()
        """
        top = ((-1)**s)*math.factorial(n-s)
        bottom = math.factorial(s)*math.factorial((0.5*(n+abs(m))-s)) * math.factorial((0.5*(n-abs(m))-s))
        temp_R = (top/bottom)*rho**(n-(2*s))
        return temp_R
    
    def get_R(self, n, m, rho):
        """
        finds the radial polynomial.  Completes the summation.
        """
        top_bound = (n - abs(m))/2
        check_int = int(top_bound)
        if not top_bound == check_int:
            raise Exception("m is not a valid number.  Must be -n, -n+2, -n+4,..., n")
        top_bound = check_int
        total = 0
        for s in list(range(top_bound + 1)):
            total += self.R_root(n, m, rho, s)
        return total
        
    def get_Z(self, n, m, rho):
        """
        finds circle of values at all theta values at a specific 
            rho value Z(rho, theta) using polar coordinates.
        returns theta as self.theta_vals values and their corresponding 
            Z values as Z
        """
        theta = np.linspace(0, 2*np.pi, 100)
        self.theta_vals = theta
        
        Rho, Theta = np.meshgrid(rho_values, theta)
        
        N = self.get_N(n, m)
        R = self.get_R(n, m, rho)
        # zernike equation depends on if m is positive or negative.
        if m >= 0:
            Z = N * R * np.cos(m * Theta)
        else:
            Z = -N * R * np.sin(m * Theta)
        return Z
    
    def surface_plot_Z(self, n, m):
        rho_vals = np.linspace(0, 1, 100)
        
        # to be honest, not sure what meshgrid does.  
        #It's just necessary for 3d plotting.
        Rho, Theta = np.meshgrid(rho_vals, self.theta_vals)
        #convert to cartesian coordinates
        X = Rho*np.cos(Theta)
        Y = Rho*np.sin(Theta)
        
        N = self.get_N(n, m)
        R = self.get_R(n, m, Rho)
        
        if m >= 0:
            Z = N * R * np.cos(m * Theta)
        else:
            Z = -N * R * np.sin(m * Theta)
        
        fig = plt.figure()
        ax = Axes3D(fig)
        ax.plot_surface(X,Y,Z, rstride=1, cstride=1, cmap='GnBu')
        
    def contour_plot_Z(self, n, m):
        rho_vals = np.linspace(0, 1, 100)
        
        # to be honest, not sure what meshgrid does.  
        #It's just necessary for 3d plotting.
        Rho, Theta = np.meshgrid(rho_vals, self.theta_vals)
        #convert to cartesian coordinates
        X = Rho*np.cos(Theta)
        Y = Rho*np.sin(Theta)
        
        N = self.get_N(n, m)
        R = self.get_R(n, m, Rho)
        
        if m >= 0:
            Z = N * R * np.cos(m * Theta)
        else:
            Z = -N * R * np.sin(m * Theta)
        V = np.linspace(np.amin(Z),np.amax(Z),100)
        plt.contourf(X,Y,Z,V, zdir='z', cmap='GnBu')
        plt.colorbar()
```
