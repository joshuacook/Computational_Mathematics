```python
f = lambda x: np.cos(x)
def cobweb(f,x0,n,xmin,xmax,ymin,ymax):
    plt.axhline(linewidth=1.0, color="black")
    plt.axvline(linewidth=1.0, color="black")
    plt.ylim((ymin,ymax))
    indep = np.linspace(xmin,xmax,5000)
    diag = lambda x: x
    plt.plot(indep,f(indep),'b')
    plt.plot(indep,diag(indep),'black')
    y0 = f(x0)
    for i in range(n):
        plt.hlines(y0, x0, y0,'r')
        x0 = y0
        y0 = f(x0)
        plt.vlines(x0, x0,y0,'r')
    

    return 0

cobweb(f,1,200,0,1.5,0,1)
```

### Analysis of Cobweb plots
The logistic map descends into chaos as r passes the value of 3.6. It is useful, however to look at what happens in between 3.8 and 3.85. If we look at behaviour prior to 3.6, the cobweb iterates toward a stable fixed point, an element of the domain of the function that is mapped to by itself. It is thus a solution to $f(x)-x=0$. Passed 3.6, however, there is an unstable fixed point, meaning that the iteration diverges from the fixed point. 

An interesting thing occurs as the value of $r$ passes near 3.85. Around this point, an "orbit" is created. Here, the cobweb neither converges to a fixed point, nor diverges infinitely. Rather it maintains a simple path around the fixed point. I am reminded of a story one of my Physics professors once told me about an exhibit at a science museum in San Francisco designed to simulate a black hole. It was a funnel shaped well into which you could toss marbles. Most of the marbles would quickly spin into the well. But if you caught it just right, you could get a marble to spin around and around the well without descending for quite some time. This being the real world and allowing for subtle imperfections on the metal making the well, the orbit of the marbles would eventually decay and fall into the well, but in theory (all things being a frictionless sphere in a vacuum in Philadelphia), the marble could spin around that well forever.
