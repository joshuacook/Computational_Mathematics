Write a python function `cobweb(f, x0, n, xmin, xmax, ymin, ymax)` to draw a cobweb plot of fixed point iteration, where
~~~
--------- --------------------------------------
 `f`      name of function to plot
 `x0`       starting point value of x
 `n`        number of fixed point iterations
 `xmin`     minimum value for $x$-axis on plot  
 `xmax`     maximum value for $x$-axis on plot  
 `ymin`     minimum value for $y$-axis on plot

This is a reminder of what a typical cobweb plot looks like (not for the function you are working with though, so your cobweb will look different).

Demonstrate that your cobweb plot works with the function call
                        
    cobweb(cos, 1.0, 200,0, 1.5, 0, 1)


```python
f = lambda x: np.cos(x)
def cobweb(f,x0,n,xmin,xmax,ymin,ymax):
    plt.axhline(linewidth=1.0, color="black")
    plt.axvline(linewidth=1.0, color="black")
    plt.ylim((ymin,ymax))
    indep = np.linspace(xmin,xmax,5000)
    diag = lambda x: x
    plt.plot(indep,f(indep),'b')
    plt.plot(indep,diag(indep),'black')
    y0 = f(x0)
    for i in range(n):
        plt.hlines(y0, x0, y0,'r')
        x0 = y0
        y0 = f(x0)
        plt.vlines(x0, x0,y0,'r')
    

    return 0

cobweb(f,1,200,0,1.5,0,1)
```

**Problem 3**

The Fourier series of a function $f(x)$ on an interval $[0,P]$ is given by

$$F(x)\sim\frac{a_0}{2}+\sum_{k=1}^n\bigg[a_n\cos\bigg(\frac{2\pi k x}{P}\bigg)+b_n\sin\bigg(\frac{2\pi k x}{P}\bigg)\bigg]$$

where 

$$a_0=\frac{2}{P}\int_0^P f(x)\ dx$$

$$a_k=\frac{2}{P}\int_0^P f(x)\cos\bigg(\frac{2\pi k x}{P}\bigg)\ dx$$

$$b_k=\frac{2}{P}\int_0^P f(x)\sin\bigg(\frac{2\pi k x}{P}\bigg)\ dx$$

a) Write a python function `fourier_coefficients(f,P,n)` that will return two lists `a` and `b` containing the coefficients $[a_0,a_1,\dots,a_n]$ and $[b_0,b_1,\dots,b_n]$ respectively.

You may use the following code to find the integral $\int_{\text{xmin}}^{\text{xmax}}f(x)\ dx:$

```python
from scipy.integrate import quad
def trapez_integration(f, xmin,xmax, n=1000):
       h=(float(xmax)-float(xmin))/float(n)
       xvals = [xmin+i*h for i in range(n+1)]
       fvals = [f(x) for x in xvals]
       integral=0.5*h*sum( [p+q for p,q in zip(fvals[1:], fvals[:-1])] )
       return integral
```

```python
def fourier_coefficients(f,P,n):
    a = np.zeros(n)
    b = np.zeros(n)
    b[0] = 0.
    a[0] = (1./P)*trapez_integration(f,-P,P)
    if (abs(a[0])<1E-10): a[0]=0
    for i in range(n-1):
        n = i+1
        a[n] = (1./P)*trapez_integration(lambda x: f(x)*np.cos(math.pi*n*x/P),-P,P)
        if (abs(a[n])<1E-10): a[n]=0
        b[n] = (1./P)*trapez_integration(lambda x: f(x)*np.sin(math.pi*n*x/P),-P,P)
        if (abs(b[n])<1E-10): b[n]=0
    return a,b

f = lambda x: x
[a,b] = fourier_coefficients(f,1,10)
```

(b) Using the function $f(x)=1-x$ and $P=1$, plot the Fourier series for $F(x)$ on the interval $[-1,2]$ using $n=5,10,25,100$. That do you observe as $n$ becomes larger? The squiggles that remain near the discontinuities are called Gibbs' Phenomenon. 

```python
def fourier_plots(func,P,n,xmin,xmax,ymin,ymax,lbl):
    plt.axhline(linewidth=1.0, color="black")
    plt.axvline(linewidth=1.0, color="black")
    plt.ylim((ymin,ymax))
    indep = np.linspace(xmin,xmax,1000)
    a,b = fourier_coefficients(func,P,n)
    dep = a[0]/2*np.ones(1000)
    for i in range(n-1):
        dep = dep + a[i+1]*np.cos(math.pi*(i+1)*indep/P)
        dep = dep + b[i+1]*np.sin(math.pi*(i+1)*indep/P)
    plt.plot(indep,dep,label=lbl)
    legend = plt.legend(loc='lower left', shadow=True)
```

```python
f = lambda x: 1-x
fourier_plots(f,1,5,-1,2,-1,2,"$n=5$")
fourier_plots(f,1,10,-1,2,-1,2,"$n=10$")
fourier_plots(f,1,25,-1,2,-1,2,"$n=25$")
fourier_plots(f,1,100,-1,2,-1,2,"$n=100$")
```
