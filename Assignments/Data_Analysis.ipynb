```python
# import libraries essential for the project
import numpy as np
import numpy.linalg as la
import math
import matplotlib.pyplot as plt
%matplotlib inline

# disable a warning that arises during the use of polyval 
import warnings
warnings.simplefilter('ignore', np.RankWarning)
```

## Import Data

```python
def import_pigs():# Import Data
    data = np.genfromtxt('pigs.csv', delimiter=',')    
    return data
```

## Take a sampling of the data

```python
def sample_data(data,samp):
    n = data[:,0].size
    choices = np.sort(np.random.choice(range(n),samp, replace=False))
    troof = np.zeros(n)
    for i in range(choices.size):
        troof[choices[i]] = 1.


    data_1 = np.zeros((samp,2))
    data_2 = np.zeros((n-samp,2))
    j = 0
    k = 0
    for i in range(n):
        if troof[i] == 1:
            data_1[j] = data[i]
            j += 1
        elif troof[i] == 0:
            data_2[k] = data[i]
            k += 1
            
    return data_1, data_2

sample_data(import_pigs(),94)
```

## Create a array of fit data
The array contains `max_degree` vectors each of the same length as the vectors in `remainder_data`. 

```python
def create_fits(max_degree,sample_data,remainder_data):
    remainder_size = remainder_data[:,0].size
    fits = np.zeros((remainder_size,max_degree+1))
    fits[:,0] = remainder_data[:,0]
    for i in range(max_degree-1):
        i = i + 2
        coefficients = np.polyfit(sample_data[:,0], sample_data[:,1], i)
        for x in range(remainder_size):   
            fits[x,i] = np.polyval(coefficients, fits[x,0]) 
    return fits 
```

## Plot data with fits
Data was plotted with a random sampling of fits in order to determine the optimal sample size to achieve a good fit. 

```python
def plotwithfit(deg,sample_data,remainder_data):
    indep = remainder_data[:,0]
    dep_1 = remainder_data[:,1]
    dep_2 = np.polyval(np.polyfit(sample_data[:,0], sample_data[:,1], deg), indep)
    plt.plot(indep,dep_1)
    plt.plot(indep,dep_2)
    plt.ylim((0,125000))
    plt.show()
    
def plotwithmultifit(deg,remainder_data,fits):
    indep_1 = remainder_data[:,0]
    indep_2 = fits[:,0]
    dep_1 = remainder_data[:,1]
    plt.figure(figsize=(20,4))
    for j in range(4):
        dep_2 = fits[:,deg+j*3]
        dep_3 = fits[:,deg+1+j*3]
        dep_4 = fits[:,deg+2+j*3]
        plt.subplot(1, 4, j+1)
        plt.plot(indep_1,dep_1, label="pigs")
        plt.plot(indep_2,dep_2, "-*", label="n="+str(deg+j*3))
        plt.plot(indep_2,dep_3, "-s", label="n="+str(deg+1+j*3))
        plt.plot(indep_2,dep_4, "-o", label="n="+str(deg+2+j*3))
        plt.ylim((0,150000))x
    plt.show()  
    
```

```python
sample_size = 94
data = import_pigs()

[sample,remainder] = sample_data(data,sample_size)

fits = create_fits(25,sample,sample)

plotwithmultifit(8,sample,fits)
```

(d) Using the 25 polynomial fits you found in the previous step (or in the same loop as you are finding the fits) calculate the root-mean-squared error between the predicted and actual $y$ value using the second data set, that is, the data that you did not use to find the fits. You will get one RMS number for each fit, the RMS error for degree $j$ is

$$\text{error}(j) = \sqrt{\frac{1}{n}\sum_{i=0}^{n-1}(y_{\text{data},i}-y_{\text{polyval},i})^2}$$

where $y_{\text{polyval},i}$ is the polynomial fit for degree $i$.

```python
def error(deg,remainder_data,fits):
    m = remainder_data[:,0].size
    yfit = fits[:,deg]
    sum = 0.
    for i in range(10,m-10):
        y_data = remainder_data[i,1] 
        sum = sum + (y_data - yfit[i])**2     
    return math.sqrt((1./(m-20))*sum)  
```

(e) Make a plot of error($j$) vs $j$ for $j = 1, 2, \dots , 25$ Based on your results, determine the optimum degree fit.

Note: because of the randomness, you will get different answers if you run the program multiple times. Before you make any conclusions you will want to run your programs several times.

```python
def rmsbydegree(samp,max_degree,iterations):
    data = import_data('pigs.csv')

    indep = np.linspace(3,26,24)
    dep = np.linspace(1,24,24)
    plt.figure(figsize=(20,4))
    for j in range(iterations):
        plt.subplot(1, 4, j+1)
        [sample,remainder] = sample_data(data,samp)
        fits = create_fits(max_degree,sample,remainder)
        for i in range(max_degree-1):
            i = i + 2
            dep[i-2] = error(i,remainder,fits)
        plt.title("optimal degree: "+str(np.argmin(dep)+2))
        plt.plot(indep,dep)
```
