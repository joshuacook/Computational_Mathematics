# the Lorenz Oscillator

The Lorenz Oscillator is described by the equations

\begin{align*}
\frac{dx}{dt}&=\sigma(y(t)-x(t))\\
\frac{dy}{dt}&=\rho x(t)-x(t)z(t)-y(t)\\
\frac{dz}{dt}&=x(t)y(t)-\beta z(t)
\end{align*}

where $\beta = 8.0/3.0$, $\sigma = 10.0$, and $\rho = 28.0$ are constants.

```python
import scipy, numpy as np
from scipy.integrate import odeint as odeint
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.colors import cnames
from matplotlib import animation
%matplotlib inline 
def lorenz_oscillator((x,y,z),t0,sigma=10.0,beta=8./3.,rho=28.0):
    return [sigma*(y - x), rho*x - x*z - y, x*y - beta*z]

r0 = [0.,1.,0.]
t = np.linspace(0,50,50000)
r_t = odeint(lorenz_oscillator, r0, t)

```

```python
# beta = 8.0/3.0
# sigma = 10.0
# rho = 28.0 


X = scipy.integrate.odeint(lorenz_oscillator, r0,t)
plt.plot(t, X[:,0], label ="$x(t)$")
plt.plot(t, X[:,1], label ="$y(t)$")
plt.plot(t, X[:,2], label ="$z(t)$")
plt.legend()
plt.show()

```
