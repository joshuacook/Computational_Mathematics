## Problem 1
Write a function named `DivDiffs` that computes the table of divided differences of the data $(x_0,y_0),(x_1,y_1),\dots,(x_n,y_n)$.

The function should take as input two 1d arrays of size $n$, and return a 2d array of size $n\times n+1$ whose first two columns hold the data and the rest of the columns the divided differences.



### Define Divided Difference Table Builder

```python
def DivDiffs(x_data, y_data, tbl=1):
    '''ordinates, abscissas, and whether or not you want a table printed, yes=1 is default'''
    n = len(x_data)
    DD = np.zeros([n,n+1])
    
    DD[:,0] = x_data
    DD[:,1] = y_data
    
    for column in range(2,n+1):
        for row in range(0, n+1-column):
             DD[row,column] = (DD[row+1,column-1] - DD[row,column-1])/(DD[row+column-1,0] - DD[row,0])
    
    # This prints the constructed divided difference table by default. 
    if tbl == 1:
        headers = ["ind", "func"]
        for i in range(DD[0].size-2):
            headers.append(str(i)+" order")
        print tabulate(DD, headers, tablefmt="latex" )
    
    return DD
    
x_1 = np.array([-2.,-1,0,1,1.5,2,2.5,3,4])
y_1 = np.array([-8.,-1,0,1,3.375,8,15.625,27,64])

dd_1 = DivDiffs(x_1,y_1,1)    
```

## Problem 2
Write a function named `InterpolantDD` that task as input the table of divided differences returned by `DivDiffs` and a list of $x$-values (within the interval delimited by the $x_i$'s and returns the values computed by the interpolating polynomial with the result from those divided differences at those $x$-values. 

$$f(x)\approx f[x_0]+(x-x_0)f[x_0,x_1]+(x-x_0)(x-x_1)f[x_0,x_1,x_2]+\dots$$

~~~
for n = 3
f(x) = DD[0,1] + (x-DD[0,0])*DD[0,2] + (x-DD[0,0])*(x-DD[1,0])*DD[0,3]
~~~

```python
def InterpolantDD(DD,x):
    interpolated_values = np.zeros(x.size)
    n = len(DD[:,0])
    monomial = 1.0

    for m in range(x.size):
        evaluated_polynomial = 0
        
        for i in range(n):
            if i == 0:
                coef = 1.0
            if i == 1:
                coef = (x[m]-DD[0,0])
            if i > 1:
                coef = 1.0
                for j in range(i):                    
                    coef = coef*(x[m]-DD[j,0])            
            evaluated_polynomial = evaluated_polynomial + coef*DD[0,i+1]# nth term

        interpolated_values[m] = evaluated_polynomial
        
    return interpolated_values

x_intp = np.array([-2.3,-1.1,1,.21,.3,1.3,2.7,3.5,4])
y_intp = InterpolantDD(dd_1,x_intp)

intp = np.column_stack([x_intp,y_intp,x_1,y_1])
print tabulate(intp,("x_intp","y_intp", "x_1","y_1"),tablefmt="latex")
plt.plot(x_intp,y_intp,"r")
plt.plot(x_1,y_1)
```
