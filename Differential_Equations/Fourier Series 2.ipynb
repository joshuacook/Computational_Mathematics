# The Fourier Representation of a function

$$f(x)\sim\frac{a_0}{2}+\sum_{k=1}^{\infty}\left[a_k\cos kx+b_k\sin kx\right]$$

## the rescaled $L^2$ inner product

$$\innerproduct{f}{g}=\frac{1}{\pi}\int_{-\pi}^{\pi}f(x)g(x)\ dx$$

## the rescaled $L^2$ norm

$$\norm{f}=\sqrt{\innerproduct{f}{f}}$$

```python
import numpy as np
import scipy.linalg.blas as blas
import matplotlib.pyplot as plt
%precision 6
%matplotlib inline
```

Under the rescaled $L^2$ inner product norm

$$\innerproduct{\cos kx}{cos lx}=\innerproduct{\sin kx}{\sin lx}=0,\ \ \forall\ \  k\neq l$$
$$\innerproduct{\cos kx}{sin lx}=0,\ \ \forall\ \  k, l$$
$$\norm{1}=\sqrt{2}$$
$$\norm{\cos kx}=\norm{\sin kx}=1$$

```python
n    = 300.
h    = 1./n
X    = np.linspace(-np.pi,np.pi,n)
ONE  = np.ones(n)
ZERO = np.zeros(n) 
dx   = 2*np.pi*h

def L2_inner(f,g,dx):
    return blas.ddot(f,g)*dx/np.pi

f = lambda x: np.exp(x)
g = lambda x: x
h = lambda x: np.exp(-x**2)
F = f(X)
G = g(X)
H = h(X)
```

```python
def fourier_series(f,k,dx):
    a_0   = L2_inner(f,ONE,dx)
    FRSRS = ZERO + a_0/2*ONE
    for i in range(k):
        i     = i+1
        a_i   = L2_inner(f,np.cos(i*X),dx)
        b_i   = L2_inner(f,np.sin(i*X),dx)
        FRSRS += a_i*np.cos(i*X)+b_i*np.sin(i*X)

    return FRSRS    
    

plt.plot(X,F)
plt.plot(X,fourier_series(F,3,dx))
```
