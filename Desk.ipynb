## Definitions
\begin{align*}
h(x)&=f(x)*g(x)=\int_{-\pi}^{\pi}f(x-y)g(y)dy\\
f(x)&=\sum_{k=-\infty}^\infty c_k\exp(ikx)\\
c_k&=\int_{-\pi}^{\pi}f(x)\exp(-ikx)dx\\
g(x)&=\sum_{k=-\infty}^\infty d_k\exp(ikx)\\
d_k&=\int_{-\pi}^{\pi}g(x)\exp(-ikx)dx\\
h(x)&=\sum_{k=-\infty}^\infty e_k\exp(ikx)
\end{align*}

## Part (a)

Find $e_k$ the complex coefficients for $h(x)$.

\begin{align*}
h(x)&=\int_{-\pi}^{\pi}f(x-y)g(y)dy\\
&=\int_{-\pi}^{\pi}\sum_{k=-\infty}^\infty c_k\exp(ik(x-y))\sum_{j=-\infty}^\infty d_j\exp(ijy)dy\\
\end{align*}

Consider a single multiple from the polynomial multiplication, $k=n,j=m\in\Z$:

\begin{align*}
I&=\int_{-\pi}^{\pi} c_m\exp(im(x-y))d_n\exp(iny)dy\\
&=c_md_n\exp(imx)\int_{-\pi}^{\pi} \exp(iny-imy)dy\\
&=c_md_n\exp(imx)\int_{-\pi}^{\pi} \exp(iy(n-m))dy
\end{align*}

\pagebreak

Case 1: $n=m$
\begin{align*}
I&=c_md_n\exp(imx)\int_{-\pi}^{\pi} dy\\
&=2\pi c_md_n\exp(imx)
\end{align*}

Case 2: $n\neq m$

\begin{align*}
I&=c_md_n\exp(imx)\int_{-\pi}^{\pi} dy\\
&=c_md_n\exp(imx)\int_{-\pi}^{\pi} \exp(iy(n-m))dy\\
&=\frac{c_md_n\exp(imx)}{i(n-m)}\exp(iy(n-m))\bigg|_{-\pi}^{\pi}\\
&=\frac{c_md_n\exp(imx)}{i(n-m)}\bigP{\exp(i\pi(n-m))-\exp(-i\pi(n-m))}=0\\
\end{align*}

Therefore, 


\begin{align*}
h(x)&=\int_{-\pi}^{\pi}\sum_{k=-\infty}^\infty c_kd_k\exp(ik(x-y)+iky)dy\\
&=\sum_{k=-\infty}^\infty c_kd_k\exp(ikx)\int_{-\pi}^{\pi}dy\\
&=2\pi\sum_{k=-\infty}^\infty c_kd_k\exp(ikx)
\end{align*}

And, 

$$e_k=2\pi c_kd_k$$

## Part (b)

For $\cos 3x$, $a_3=1$ with $b_k=0$ and $a_{k\neq3}=0$. Therefore, $c_3=\frac{1}{2}$ and $c_{-3}=\frac{1}{2}$ with all $c_{k\neq 3,-3}=0$.

For $\sin 2x$, $b_2=1$ with $a_k=0$ and $b_{k\neq2}=0$. Therefore, $d_2=\frac{-i}{2}$ and $d_{-2}=\frac{i}{2}$ with all $d_{k\neq 2,-2}=0$.

All $e_{k\neq-3,-2,2,3}$ will trivially be 0. 

\begin{align*}
e_{-3}&=\frac{1}{2}\cdot0=0\\
e_{-2}&=0\cdot\frac{i}{2}=0\\
e_2&=0\cdot\frac{-i}{2}=0\\
e_3&=\frac{1}{2}\cdot0=0
\end{align*}

# Derivation of the Momentum Operator

\begin{align*}
\left\langle \hat{p}_x \right\rangle &= \left\langle \hbar\hat{k} \right\rangle\\
&= \hbar \left\langle \psi(k)\left|\hat{k}\right|\psi(k) \right\rangle\\
&= \hbar \int_{-\infty}^\infty\psi^*(k)k\psi(k)dk\\
&= \hbar \int_{-\infty}^\infty \frac{1}{\sqrt{2\pi}}\int_{-\infty}^\infty \psi^*(x')e^{-ikx'} k\frac{1}{\sqrt{2\pi}}\int_{-\infty}^\infty \psi(x)e^{ikx}dxdx'dk\\
&= \frac{\hbar}{2\pi} \int_{-\infty}^\infty \psi^*(x')\int_{-\infty}^\infty e^{-ikx'} k\int_{-\infty}^\infty \psi(x)e^{ikx}dxdkdx'\\
&= -i\frac{\hbar}{2\pi} \int_{-\infty}^\infty \psi^*(x')\int_{-\infty}^\infty e^{-ikx'} \int_{-\infty}^\infty \psi(x)ike^{ikx}dxdkdx'\\
&= -i\frac{\hbar}{2\pi} \int_{-\infty}^\infty \psi^*(x')\int_{-\infty}^\infty e^{-ikx'} \int_{-\infty}^\infty \psi(x)\frac{d}{dx}e^{ikx}dxdkdx'\\
\end{align*}

Consider for a moment

\begin{align*}
\int_{-\infty}^\infty \psi(x)\frac{d}{dx}e^{ikx}dx&=\psi(x)e^{ikx}\bigg|_{-\infty}^\infty-\int_{-\infty}^\infty \frac{d}{dx}\left(\psi(x)\right)e^{ikx}dx\\
\end{align*}

For the square integrable functions $\psi{x}$ and $e^{ikx}$

\begin{align*}\psi(x)e^{ikx}\bigg|_{-\infty}^\infty&= 0
\end{align*}


\begin{align*}
\int_{-\infty}^\infty \psi(x)\frac{d}{dx}e^{ikx}dx&=-\int_{-\infty}^\infty e^{ikx}\frac{d}{dx}\psi(x)dx\\
\end{align*}

Then,

\begin{align*}
\left\langle \hat{p}_x \right\rangle 
&= i\frac{\hbar}{2\pi} \int_{-\infty}^\infty \psi^*(x')\int_{-\infty}^\infty e^{-ikx'} \int_{-\infty}^\infty e^{ikx}\frac{d}{dx}\psi(x)dxdkdx'\\
&= i\frac{\hbar}{2\pi} \int_{-\infty}^\infty \int_{-\infty}^\infty\int_{-\infty}^\infty \psi^*(x') e^{ik(x-x')}  \frac{d}{dx}\psi(x)dxdkdx'\\
\end{align*}

The Dirac Delta Measure gives 

\begin{align*}
\int_{-\infty}^\infty e^{ik(x-x')} dk &=2\pi\delta(x-x')
\end{align*}

So,

\begin{align*}
\left\langle \hat{p}_x \right\rangle &= -i\hbar  \int_{-\infty}^\infty\int_{-\infty}^\infty \psi^*(x') \delta(x'-x)  \frac{d}{dx}\psi(x)dxdx'\\
&=  \int_{-\infty}^\infty \psi^*(x) \left( -i\hbar  \frac{d}{dx}\right)\psi(x)dx\\
&= \left\langle \psi^*(x)\left|\left( -i\hbar  \frac{d}{dx}\right)\right|\psi(x)\right\rangle
\end{align*}

And therefore, $p_x=-i\hbar  \frac{d}{dx}$.
